## 1.8 정렬하기

```python
# df_null_count 데이터프레임에 있는 결측치 수 컬럼을 sort_values를 통해서 정렬
# 결측치가 많은 순으로 상위 10개만 출력
df_null_count.sort_values(by='결측치수',ascending=False).head(10)
```

### 1.12.3 텍스트 데이터

```python
# 색인하기 전에 상호명 중에 종합병원이 아닌 데이터를 찾아봅니다.
df_seoul_hospital[~df_seoul_hospital['상호명'].str.contains('종합병원')]
```

```python
df_seoul_hospital.loc[~df_seoul_hospital['상호명'].str.contains('종합병원'),'상호명'].unique()
```

```python
# 상호명에서 특정 단어가 들어가는 데이터만 가져옵니다. - 꽃배달
df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달')]
```

```python
# 특정 단어가 들어가는 데이터만 가져옵니다. - 의료기
df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('의료기')]
```

```python
# '꽃배달|의료기|장례식장|상담소|어린이집'은 종합병원과 무관하기 때문에
# 전처리를 위해 해당 텍스트를 한번에 검색합니다.
# 제거할 데이터의 인덱스만 drop_row에 담아주고 list형태로 변환합니다.
df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달|의료기|장례식장|상담소|어린이집')]
```

```python
drop_row = df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달|의료기|장례식장|상담소|어린이집')].index
drop_row = drop_row.tolist()
drop_row
```

```python
# 의원으로 끝나는 데이터도 종합병원으로 불 수 없기 때문에 인덱스를 찾아서
# drop_row2에 담아주고 list 형태로 반환합니다
drop_row2 = df_seoul_hospital[df_seoul_hospital['상호명'].str.endswith('의원')].index
drop_row2 = drop_row2.tolist()
drop_row2
```

```python
# 삭제할 행을 drop_row에 합쳐줍니다
drop_row = drop_row + drop_row2
len(drop_row)
```

```python
# 해당 셀을 삭제하고 삭제 전과 후의 행의 갯수를 비교합니다.
print('전처리 전' , df_seoul_hospital.shape)
df_seoul_hospital.drop(drop_row,axis=0)
print('전처리 후', df_seoul_hospital.shape)
```

```python
# 시군구명에 따라 종합병원의 숫자를 countplot으로 그립니다.
c = df_seoul_hospital['시군구명'].value_counts()
c.plot.bar()
```

### 1.13.1 Folium 사용예제

```python
# 지도의 중심을 지정하기 위해 위도와 경도의 평균을 구합니다.
print('위도의 평균값 : ',df_seoul_hospital['위도'].mean())
print('경도의 평균값 : ',df_seoul_hospital['경도'].mean())
```

```python
map = folium.Map(location=[df_seoul_hospital['위도'].mean(),df_seoul_hospital['경도'].mean()],zoom_start=10)
for n in df_seoul_hospital.index:
    name =df_seoul_hospital.loc[n,'상호명']
    address = df_seoul_hospital.loc[n,'도로명주소']
    popup = f"{name}-{address}"
    location = [df_seoul_hospital.loc[n,'위도'],df_seoul_hospital.loc[n,'경도']]
    folium.Marker(
            location = location,
            popup = popup
    ).add_to(map)
map
```
